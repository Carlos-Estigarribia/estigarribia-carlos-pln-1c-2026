## 1. El problema: para la computadora, el texto todavía es texto plano

En esta primera parte vamos a mirar qué pasa cuando tratamos un texto solo como una secuencia de caracteres.


```python
texto = "El gobernador de la provincia de Buenos Aires, Axel Kicilof viajó a Europa para buscar inversiones y apoyos políticos a su candidatura a presidente de Argentina."
print(texto)
```

Antes de seguir, hace una prueba breve:
- calcula `len(texto)`;
- compara ese valor con la salida de `texto.split()`;
- preguntate qué información lingüística falta.

Idea central: separar por espacios no equivale a comprender el lenguaje.


```python
len(texto) # calcula la cantidad de caracteres
```

```python
palabras = texto.split()
print(palabras) 
```

```python
len(palabras) # calcula la cantidad de palabras. La info que falta es importancia o jerarquía de las palabras
```

### Por qué necesitamos algo mejor

`split()` resuelve una parte del problema, pero deja preguntas abiertas:
- `"ACME,"` y `"Chang,"` conservan la coma. No son palabras limpias.
- `"La"` y `"los"` son artículos, mientras que `"viajara"` es una forma conjugada del verbo `"viajar"`. Cómo llevás cada palabra a su forma base para analizar frecuencias o temas?
- Cómo identifica la máquina que `"ACME"` es una organización y que `"Lisa Chang"` es una persona?

**Pausa breve de observación:** si un programa solo ve caracteres, todavía no puede distinguir personas, organizaciones, lugares ni relaciones gramaticales.


## 2. La solución: spaCy como intérprete lingüístico

En este punto pasamos de texto plano a texto enriquecido lingüísticamente. `spaCy` toma una cadena y devuelve un objeto `Doc` con tokens, lemas, categorías gramaticales y otra información útil para el análisis.


```python
!pip install spacy -q
!python -m spacy download es_core_news_sm -q
```

```python
import spacy
```

```python
# Cargamos el modelo de español
nlp = spacy.load("es_core_news_sm") # nlp es la variable del spacy para el español

# Procesamos el texto
doc = nlp(texto)      # con nuestro nlp procesamos el texto y lo guardamos en la variable doc

print(doc)
```

## 3. Qué información agrega spaCy al texto

A partir de `doc`, ya no trabajamos solo con caracteres: trabajamos con unidades lingüísticas y con atributos asociados a cada una.


**Tokenización: la unidad mínima de análisis**

Un token es una unidad mínima de texto con significado, como una palabra, un número o un signo de puntuación.


```python
for token in doc:
    print(token.text)
```

**Lematización: la forma base de cada palabra**

El lema es la forma canónica o de diccionario de una palabra. Permite agrupar variantes que comparten significado.


```python
print("PALABRA		LEMA")
print("-------		----")
for token in doc:
    print(f"{token.text:<15}	{token.lemma_}")
```

**Etiquetado gramatical (POS): el rol de cada palabra**

Consiste en asignar a cada palabra su función gramatical, por ejemplo sustantivo, verbo o adjetivo.


```python
print("PALABRA		POS		EXPLICACIÓN")
print("-------		---		-----------")
for token in doc:
    print(f"{token.text:<15}	{token.pos_:<10}	{spacy.explain(token.pos_)}")
```
