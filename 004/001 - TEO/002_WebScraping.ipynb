```python
# 2. Importamos las bibliotecas
import requests
from bs4 import BeautifulSoup
```

```python
# Vamos primero con el proyecto Gutemberg
#url = "https://www.gutemberg.org/cache/epub/58221/pg58221-images.html" # No la carga

url = "https://www.gutenberg.org/files/58221/58221-h/58221-h.htm"
# hacemos el pedido Get (pedido de información) al servidor web
pagina = requests.get(url)

# Importante! Revisamos si el pedido fue exitoso (codigo 200). Evita procesar si tira error.
pagina.raise_for_status()

# Extraemos el texto crudo devuelto por la petición
contenido = pagina.text
```

```python
# Con el texto descargado
# 5. Creamos el objeto BeautifulSoup para analizar el HTML
soup = BeautifulSoup(contenido, 'html.parser')
```

```python
# Método find_all()
# Cuando buscamos información masiva (como la totalidad de los parrafos de un documento), utilizamos find_all()
# Este metodo devuelve un objeto iterable (como una lista) que contiene cada coincidencia existente en todo el arbol HTML
parrafos = soup.find_all("p")
print(f"Encontramos {len(parrafos)} párrafos")
```

```python
h1 = soup.find_all("h1")
print("Encabezado de la página:", h1[0].text if h1 else "No encontrado")
```

STackoverflow

```python
# 3. Definimos la página web que queremos analizar
url = "https://stackoverflow.com/questions/415511/how-to-get-the-current-time-in-python"
```

```python
# 4. Obtenemos el contenido de la página web
pagina = requests.get(url)
contenido = pagina.text
```

```python
# 5. Creamos el objeto BeautifulSoup para analizar el HTML
soup = BeautifulSoup(contenido, 'html.parser')
```

```python
# 6. Extraemos información básica
# Obtenemos el título de la página
print("TÍTULO DE LA PÁGINA:")
print(soup.title.text)
print("-" * 50)
```

```python
# 7. Extraemos la pregunta
print("\nPREGUNTA:")
pregunta = soup.find("div", {"class": "question"})
texto_pregunta = pregunta.find("div", {"class": "s-prose js-post-body"})
print(texto_pregunta.get_text().strip())
print("-" * 50)
```

```python
# 8. Extraemos la mejor respuesta
print("\nMEJOR RESPUESTA:")
respuesta = soup.find("div", {"class": "answer"})
texto_respuesta = respuesta.find("div", {"class": "s-prose js-post-body"})
print(texto_respuesta.get_text().strip())
```
